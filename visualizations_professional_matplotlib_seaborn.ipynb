```python
# histogram
fig, ax = plt.subplots()
ax.hist(x=df_iris["sepal_width"], alpha=0.75, bins=20)
ax.text(0.057, .97, '$\\bf{Matplotlib\ Histogram\ Title}$', 
         fontsize=42, color=custom_colors['GRAY_2'], transform=fig.transFigure)
ax.text(0.06, .91, "Subtitle Text", 
         fontsize=22, color=custom_colors['GRAY_3'], transform=fig.transFigure)
plt.xlabel("sepal_width")
plt.ylabel("Number")
plt.show()
plt.close()
```

```python
# histogram
fig, ax = plt.subplots()
for each_column in df_iris.columns[0:4]:
    ax.hist(x=df_iris[each_column], alpha=0.8, bins=10)
ax.legend(df_iris.columns, loc="upper right")
ax.text(0.057, .97, '$\\bf{Matplotlib\ Histogram\ Title}$', 
         fontsize=42, color=custom_colors['GRAY_2'], transform=fig.transFigure)
ax.text(0.06, .91, "Subtitle Text", 
         fontsize=22, color=custom_colors['GRAY_3'], transform=fig.transFigure)
plt.xlabel("Value")
plt.ylabel("Number")
plt.show()
plt.close()
```

```python
# scatter
fig, ax = plt.subplots()
for each_species in df_iris['species'].unique():
    each_df = df_iris[df_iris['species'] == each_species]
    ax.scatter(x=each_df.sepal_length, y=each_df.sepal_width, alpha=0.5)
ax.legend(df_iris['species'].unique(), loc="upper right")
ax.text(0.057, .97, '$\\bf{Matplotlib\ Scatterplot\ Title}$', 
         fontsize=42, color=custom_colors['GRAY_2'], transform=fig.transFigure)
ax.text(0.06, .91, "Subtitle Text", 
         fontsize=22, color=custom_colors['GRAY_3'], transform=fig.transFigure)
plt.xlabel("sepal_length")
plt.ylabel("sepal_width")
plt.show()
plt.close()
```

```python
# barh
fig, ax = plt.subplots()
bar_data = df_iris.mean()
bar_index = np.arange(len(bar_data))
bar_height = 0.7
bars = ax.barh(y=bar_index, width=bar_data, height=bar_height)
ax.invert_yaxis()
ax.text(-0.038, .94, '$\\bf{Matplotlib\ Horizontal\ Bar\ Title}$', 
         fontsize=42, color=custom_colors['GRAY_2'], transform=fig.transFigure)
ax.text(-0.034, .88, "Subtitle Text", 
         fontsize=22, color=custom_colors['GRAY_3'], transform=fig.transFigure)
ax.spines['left'].set_visible(False)
ax.tick_params(axis='y', length=0)
for each_index, each_bar in enumerate(bars):
    each_value = np.round(df_iris.mean()[each_index], decimals=3)
    plt.text(each_bar.get_width()-0.7, each_bar.get_y()+.36, 
             each_value, fontsize=24, color="white", weight="bold", va="center")
plt.yticks(bar_index, df_iris.columns + " mean")
plt.xlabel("Value")
plt.show()
plt.close()
```

```python
# barh
fig, ax = plt.subplots()
bar_data = df_iris.mean().sort_values(ascending=False)
bar_index = np.arange(len(bar_data))
bar_height = 0.7
colors = [custom_colors['GRAY_6']] + [custom_colors['BLUE_1']] + [custom_colors['GRAY_6']] * 2
bars = ax.barh(y=bar_index, width=bar_data, height=bar_height, color=colors)
ax.invert_yaxis()
ax.text(-0.038, .94, '$\\bf{Matplotlib\ Horizontal\ Bar\ Title}$', 
         fontsize=42, color=custom_colors['GRAY_2'], transform=fig.transFigure)
ax.text(-0.034, .88, "Subtitle Text", 
         fontsize=22, color=custom_colors['GRAY_3'], transform=fig.transFigure)
ax.spines['left'].set_visible(False)
ax.tick_params(axis='y', length=0)
for each_index, each_bar in enumerate(bars):
    each_value = np.round(df_iris.mean()[each_index], decimals=3)
    plt.text(each_bar.get_width()-0.7, each_bar.get_y()+.36, 
             each_value, fontsize=24, color="white", weight="bold", va="center")
plt.yticks(bar_index, df_iris.columns + " mean")
plt.xlabel("Value")
plt.show()
plt.close()
```

```python
# grouped bar
fig, ax = plt.subplots()
bar_data = df_iris.groupby("species").mean()
bar_index = np.arange(bar_data.shape[1])
bar_width = 0.7
for each_index in range(bar_data.shape[0]):
    ax.bar(x=bar_index - bar_width/2 + (each_index / bar_data.shape[0] * bar_width),
       height=bar_data.iloc[each_index], width=bar_width/bar_data.shape[0], alpha=0.9, align='edge')
ax.legend(df_iris.species.unique(), loc="upper right")
ax.text(0.07, .97, '$\\bf{Matplotlib\ Vertical\ Bar\ Plot\ Title}$', 
         fontsize=42, color=custom_colors['GRAY_2'], transform=fig.transFigure)
ax.text(0.074, .91, "Subtitle Text", 
         fontsize=22, color=custom_colors['GRAY_3'], transform=fig.transFigure)
plt.xticks(bar_index, df_iris.columns[0:4] + " mean")
plt.xlabel("Feature")
plt.ylabel("Value")
plt.show()
plt.close()
```

```python
# grouped bar
bar_data = df_iris.groupby("species").mean().T
colors = [custom_colors['BLUE_1'], custom_colors['GRAY_7'], custom_colors['GRAY_9']]
ax = bar_data.plot(kind='bar', color=colors, width=.7)
plt.figtext(0.07, .97, '$\\bf{Matplotlib\ Vertical\ Bar\ Plot\ Title}$', 
            fontsize=42, color=custom_colors['GRAY_2'])
plt.figtext(0.074, .91, "Subtitle Text", 
            fontsize=22, color=custom_colors['GRAY_3'])
plt.legend(title=" ")
plt.xlabel("Feature")
plt.ylabel("Value")
plt.xticks(rotation=0) 
for each_index, each_bar in enumerate(ax.patches):
    if each_index in range(0,4):
        each_value = round(each_bar.get_height(), ndigits=1)
        ax.text(each_bar.get_x() + .03, each_bar.get_height() + 0.15,
                each_value, fontsize=20, color=custom_colors['GRAY_1'], weight="bold", va="center")
plt.show()
plt.close()
```

```python
# line
fig, ax = plt.subplots()
ax.plot(range(0,len(df_iris["petal_length"])), df_iris["petal_length"], '-', linewidth=4,)
subset_df_iris = df_iris.iloc[50:59]
ax.plot(range(50,59), subset_df_iris["petal_length"], 'o', markersize=12, color=custom_colors["BLUE_1"])
ax.text(0.0655, .97, '$\\bf{Matplotlib\ Line\ Plot\ Title}$', 
         fontsize=42, color=custom_colors['GRAY_2'], transform=fig.transFigure)
ax.text(0.069, .91, "Subtitle Text", 
         fontsize=22, color=custom_colors['GRAY_3'], transform=fig.transFigure)
below_line_index = [53, 57]
for each_row in subset_df_iris.itertuples():
    each_x = each_row.Index
    each_y = each_row.petal_length
    if each_x in below_line_index:
        plt.annotate(each_y, [each_x - .26, each_y - .3], 
                     weight='bold', fontsize=18, color=custom_colors['GRAY_1'])
    else:
        plt.annotate(each_y, [each_x - .26, each_y + .18], 
                     weight='bold', fontsize=18, color=custom_colors['GRAY_1'])
plt.xlabel("Index")
plt.ylabel("Value")
plt.xlim([45,59])
plt.ylim([1,6])
plt.axvline(x=49, color=custom_colors['GRAY_1'], linewidth=1)
plt.annotate("event text description", [49 + 0.5, 6 - 0.1])
plt.show()
plt.close()
```
