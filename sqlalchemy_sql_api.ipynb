```python
# IN - db.select([census.columns.state, census.columns.gender]).where(census.columns.state.in_(['Texas', 'New York']))
# AND, OR, NOT - db.select([census]).where(db.and_(census.columns.state == 'California', census.columns.gender != 'M'))
# ORDER BY - db.select([census]).order_by(db.desc(census.columns.state), census.columns.pop2000)
# FUNCTIONS (i.e. SUM) - db.select([db.func.sum(census.columns.pop2008)])
# GROUP BY - db.select([db.func.sum(census.columns.pop2008).label('pop2008'), census.columns.gender]).group_by(census.columns.gender)
# DISTINCT - db.select([census.columns.state.distinct()])
# AUTO JOINS - db.select([census.columns.pop2008, state_fact.columns.abbreviation])
# MANUAL JOINS - query = db.select([census, state_fact])
#                query = query.select_from(census.join(state_fact, census.columns.state == state_fact.columns.name))
```

# creating tables

```python
employees = db.Table('employees', metadata,
                     db.Column('Id', db.Integer()),
                     db.Column('name', db.String(255), nullable=False),
                     db.Column('salary', db.Float(), default=100.0),
                     db.Column('active', db.Boolean(), default=True))
```

```python
employees.create(engine)
```

```python
inspector = db.inspect(engine)
inspector.get_table_names()
```

```python
inspector.get_columns('employees')
```

# inserting rows

```python
employees = db.Table('employees', metadata, autoload=True, autoload_with=engine)
```

```python
# single insert
query = db.insert(employees).values(Id=1, name='james', salary=60000.00, active=True) 
result_proxy = connection.execute(query)
```

```python
# multiple insert 
query = db.insert(employees) 
values_list = [{'Id':'2', 'name':'jeff', 'salary':80000, 'active':False},
               {'Id':'3', 'name':'brad', 'salary':70000, 'active':True}]
result_proxy = connection.execute(query ,values_list)
```

```python
query = db.select([employees])
results_proxy = connection.execute(query)
```

```python
df = pd.DataFrame(results_proxy.fetchall())
df.columns = results_proxy.keys()
df
```

# update data

```python
employees = db.Table('employees', metadata, autoload=True, autoload_with=engine)
```

```python
query = db.update(employees).values(salary = 5150).where(employees.columns.Id == 1)
results_proxy = connection.execute(query)
```

```python
query = db.select([employees])
results_proxy = connection.execute(query)
```

```python
df = pd.DataFrame(results_proxy.fetchall())
df.columns = results_proxy.keys()
df
```

# delete rows

```python
query = db.delete(employees).where(employees.columns.active == False)
results_proxy = connection.execute(query)
```

```python
query = db.select([employees])
results_proxy = connection.execute(query)
```

```python
df = pd.DataFrame(results_proxy.fetchall())
df.columns = results_proxy.keys()
df
```

# dropping table

```python
# to drop all tables in dataase use: metadata.drop_all(engine)
connection.close()
employees.drop(engine)
```
