# Neo4j

# setup

```python
import pandas as pd
```

```python
from py2neo import Graph, Node, Relationship
```

```python
# connection 
graph = Graph(uri='http://localhost:7474', auth=('neo4j','123'))
```

# execute cypher 

```python
# run cypher query language
# results as dataframe
graph.run("MATCH (a:Movie) RETURN a.title AS MOVIE_TITLE LIMIT 25" ).to_data_frame()
```

```python
# run cypher query language 
# evaluate to single value
graph.evaluate("MATCH (a:Movie) RETURN count(a)")
```

# pythonic api

```python
# create node 
graph.create(Node("Movie", title="Maximun Overdrive", released="1986"))
```

```python
# query nodes
pd.DataFrame(graph.nodes.match("Movie"))
```

```python
# query node
graph.nodes.match("Movie", title='Maximun Overdrive').first()
```

```python
# query relationship
node = graph.nodes.match("Person", name="Keanu Reeves").first()
list(graph.relationships.match((node, None), "ACTED_IN") .limit(10))
```
