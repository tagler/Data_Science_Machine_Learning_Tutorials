```python
def custom_theme():
    # font
    font = "Helvetica Neue"
    font_size = 16
    font_larger_size = 18
    subtitle_font_size = 20
    title_font_size = 36
    # colors
    color_font_axes_labels = colors["GRAY_7"]
    color_title = colors["GRAY_3"]
    color_subtitle = colors["GRAY_5"]
    color_primary_mark = colors_tableau["BLUE"]
    # config
    config = {
        # background
        "background": colors["WHITE_1"],
        "config" : {
            # axis
            "axis": {
                # grid
                "grid": False,
                # axes
                "domainColor": color_font_axes_labels,
                "domainWidth" : 1.25,
                # ticks
                "tickColor" : color_font_axes_labels,
                "tickSize" : 5,
                "tickWidth" : 1.25,
                "tickCount" : 6,
                # title
                "titleFont" : font,
                "titleColor" : color_font_axes_labels,
                "titleFontSize" : font_larger_size,
                "titleFontWeight" : 400,
                "titlePadding" : 5,
                # labels
                "labelFont" : font,
                "labelColor" : color_font_axes_labels,
                "labelFontSize" : font_size,
                "labelFontWeight" : 400,
                "labelPadding" : 5,
            },
            # axis X
            "axisX": {
                # title
                "titlePadding" : 5,
            },
            # axis Y 
            "axisY": {
                # title
                "titlePadding" : 10,
            },
            # title
            "title": {
                "font": font,
                "color": color_title,
                "fontSize": title_font_size,
                "fontWeight": 600,
                # subtitle
                "subtitleFont" : font,
                "subtitleColor": color_subtitle,
                "subtitleFontSize": subtitle_font_size,
                "subtitleFontWeight" : 400,
                "subtitlePadding" : 8,
                "anchor": "start",
                "offset": 14,
            },
            # legend
            "legend": {
                # title
                "titleFont": font,
                "titleColor": color_font_axes_labels,
                "titleFontSize": font_size,
                "titleFontWeight" : 600,
                "titlePadding" : 5,
                # labels
                "labelFont": font,
                "labelColor": color_font_axes_labels,
                "labelFontSize": font_size,
                "labelFontWeight" : 400,
                "offset" : 20,
                "padding" : 5,
            },
            # facets
            "header": {
                # title
                "titleFont": font,
                "titleColor": color_font_axes_labels,
                "titleFontSize": font_larger_size,
                "titleFontWeight" : 400,
                # labels
                "labelFont": font,
                "labelColor": color_font_axes_labels,
                "labelFontSize": font_size,
                "labelFontWeight" : 400,
            },
            # color cycles
            "range": {
                "category": color_cycle_custom,
            },
            # view
            "view" : {
                # outline
                "stroke" : "transparent",
            },
            # points
            "circle" : {
                "color" : color_primary_mark,
                "size" : 100,
            },
            "point": {
               "filled": True,
            },
            # bars
            "bar" : {
                "color" : color_primary_mark,
            },
            # lines
            "line" : {
                "color" : color_primary_mark,
            },
            # text
            "text" : {
                "font" : font,
                "fontSize" : font_larger_size,
                "color" : colors["GRAY_5"],
            },
            # rule
            "rule" : {
                "size" : 1.25,
                "color" : color_font_axes_labels,
            },
            # area
            "area": {
               "fill": color_primary_mark,
            }
        }
    }
    return config
```

```python
alt.themes.register("custom_theme", custom_theme);
```

```python
alt.themes.enable("custom_theme");
```

```python
# example api syntax: alt.Chart({data_frame}).{mark_method}.encode({encodings}).interactive()
# note: prefers long-form dataframes (use .melt() to transform)
```

```python
# marks: .mark_bar(), .mark_circle(), .mark_line(), .mark_point(), .mark_boxplot(), .mark_rule(), .mark_text()
# encoding: Q(quantitative), O(ordinal, discrete ordered), N(nominal, discrete unordered), T(temporal)
# properties: color, opacity, shape, size
```

```python
# note: alt.text(format={format string}) uses D3 format codes
# https://github.com/d3/d3-format#locale_format
```

```python
# histograms
chart = alt.Chart(
    data=df_iris
).mark_bar(
).encode(
    x=alt.X(
        "sepal_width:Q", 
        bin=alt.Bin(maxbins=25)
    ),
    y='count()',
    opacity=alt.value(0.7)
).properties(
    title={
        "text" : "Altair Histogram Plot Title",  
        "subtitle" : "Subtitle Text"
    },
    width=width, 
    height=height,
).interactive()
# display
chart.display()
```

```python
# histogram 
chart = alt.Chart(
    data=df_iris
).transform_fold(
    ['sepal_length', 'sepal_width','petal_length', 'petal_width'],
    as_=['Feature', 'Value'],
).mark_bar(
    opacity=0.6,
).encode(
    x=alt.X(
        'Value:Q', 
        bin=alt.Bin(maxbins=50)
    ),
    y=alt.Y(
        'count()', 
        stack=None
    ),
    color=alt.Color(
        'Feature:N', 
        legend=alt.Legend(orient="right")
    ),
).properties(
    title={
        "text" : "Altair Multiple Histogram Plot Title",  
        "subtitle" : "Subtitle Text"
    },
    width=width, 
    height=height,
).interactive()
# display
chart.display()
```

```python
# bar plot
bar_data = pd.DataFrame(df_iris.mean()).reset_index()
bar_data.columns = ["Feature", "Mean_Value"]
# base
base = alt.Chart(
    data=bar_data
).encode(
    x=alt.X(
        "Mean_Value:Q",
        title="Mean Value"
    ),
    y=alt.Y(
        "Feature:N", 
        title=None,
        sort = alt.EncodingSortField(
            field="Mean_Value",  
            order="descending"
        )
    ),
    opacity=alt.value(0.9),
)
# bars
bars = base.mark_bar(
    size=70,
).encode(
    color= alt.condition(
        alt.datum.Feature == "petal_length", 
        alt.value(colors_tableau["BLUE"]),
        alt.value(colors["GRAY_6"]),
    )
)
# text
text = base.mark_text(
    size=28,
    color=colors["WHITE_1"],
    align='center',
    baseline='middle',
    dx=-45,
).encode(
    text=alt.Text('Mean_Value:Q', format=',.2f'),
)
# chart
chart = alt.layer(
    bars, text
).properties(
    title={
        "text" : "Altair Bar Plot Title",  
        "subtitle" : "Subtitle Text",
        "offset" : 4
    },
    width=width, 
    height=height,
).configure_axisY(
    domainOpacity=0,
    labelPadding=10,
    tickSize=0,
)
# display
chart.display()
```

```python
# bar plot 
bar_data = df_iris.groupby("species").mean()
bar_data["species"] = bar_data.index
bar_data.reset_index(drop=True, inplace=True)
bar_data_melt = bar_data.melt(
    id_vars=['species'], 
    value_vars=['sepal_length', 'sepal_width','petal_length', 'petal_width']
)
bar_data_melt.columns = ["Species", "Feature", "Value"]
# base
base = alt.Chart(
    data=bar_data_melt
).encode(
    y=alt.Y(
        "Value:Q", 
        title="Mean Value"
    ),
    x=alt.X(
        "Species:O", 
        axis=alt.Axis(
            title=None, 
            labels=False, 
            ticks=False
        )
    ),
).properties(
    width=150,
    height=410,
)
# bars
custom_bar_colors = [
    colors_tableau['BLUE'], 
    colors['GRAY_7'], 
    colors['GRAY_9']
]
bars = base.mark_bar(
    size=50,
).encode(
    color=alt.Color(
        "Species:N", 
        title="Species",
        legend=alt.Legend(orient="right"),
        scale=alt.Scale(range=custom_bar_colors)
    ),
    opacity=alt.value(0.95),
)
# text
text = base.mark_text(
    size=18,
    fontWeight=600,
    align='center',
    baseline='middle',
    dy=-10,
).encode(
    text=alt.Text(
        'Value:Q', 
        format=',.1f'
    ),
    opacity=alt.condition(
        alt.datum.Species == "setosa",
        alt.value(1), 
        alt.value(0)
    )
)
# chart
chart = alt.layer(
    bars, text, 
    data=bar_data_melt
).facet(
    column=alt.Column(
        "Feature:N",
        title="Feature", 
        header=alt.Header(
            orient="bottom", 
            titlePadding=-2, 
            labelPadding=8
        )
    ),
    title={
        "text" : "Altair Grouped Bar Plot Title",  
        "subtitle" : "Subtitle Text"
    },
).configure_scale(
    bandPaddingInner=0.25,
    bandPaddingOuter=0.1,
)
# display
chart.display()
```

```python
# line plot
line_data = pd.DataFrame(df_iris["petal_length"]).reset_index()
line_data.columns = ["index_time", "petal_length"]
line_data_subset = line_data.iloc[45:60]
# base
base = alt.Chart(
    data=line_data_subset
).encode(
    x=alt.X(
        "index_time:Q", 
        title="Index",
        scale=alt.Scale(
            zero=False, 
            clamp=True, 
            domain=(45,60)
        )
    ),
    y=alt.Y(
        "petal_length:Q", 
        title="petal_length Value",
        scale=alt.Scale(
            zero=False, 
            clamp=True, 
            domain=(1,6)
        )
    ),
)
# liness
lines = base.mark_line(
    size=5,
)
# points
points = base.mark_circle(
    size=150,
).encode(
    opacity=alt.condition(
        (
            (alt.datum.index_time >= 50) & 
            (alt.datum.index_time <= 58)
        ),
        alt.value(1), 
        alt.value(0)
    )
)
# text up
text_up = base.mark_text(
    size=20,
    fontWeight=600,
    color=colors['GRAY_1'],
    align='center',
    baseline='middle',
    dx=-1,
    dy=-22,
).encode(
    text=alt.condition(
        (
            (alt.datum.index_time != 53) & 
            (alt.datum.index_time != 57) & 
            (alt.datum.index_time > 49) & 
            (alt.datum.index_time < 59)
        ),
        alt.Text(
            'petal_length:Q', 
            format=',.1f'), 
        alt.Text()
    )
)
# text down
text_down = base.mark_text(
    size=20,
    fontWeight=600,
    color=colors['GRAY_1'],
    align='center',
    baseline='middle',
    dx=-1,
    dy=20,
).encode(
    text=alt.condition(
        (
            (alt.datum.index_time == 53) | 
            (alt.datum.index_time == 57)
        ),
        alt.Text(
            'petal_length:Q', 
            format=',.1f'), 
        alt.Text()
    )
)
# vertical line
overlay = pd.DataFrame({
    'x': [49]
})
vline = alt.Chart(
    overlay
).mark_rule(
    color=colors['GRAY_7'],
    size=1.5,
).encode(
    x='x:Q'
)
# text annotation
annotation = pd.DataFrame({
    'x' : [50], 
    'y' : [5.8], 
    'text' : ["event text description"]
})
text_annotation = alt.Chart(
    annotation
).mark_text(
    color=colors['GRAY_3'],
    dx=65,
    size=18,
).encode(
    x='x:Q',
    y='y:Q',
    text='text',
)
# chart
chart = alt.layer(
    lines, points, text_up, text_down, vline, text_annotation
).properties(
    title={
        "text" : "Altair Single Line Plot Title",  
        "subtitle" : "Subtitle Text",
        "offset" : 15
    },
    width=width, 
    height=height
)
# display
chart.display()
```

```python
# smoothing plot
df_iris_melt = df_iris.reset_index().melt(
    id_vars=['index'], 
    value_vars=['petal_length','sepal_length']
)
# base
custom_point_colors = [
    colors_tableau["BLUE"], 
    colors_tableau["RED"],
]
base = alt.Chart(
    data=df_iris_melt
).encode(
    x=alt.X(
        'index', 
        title="Index", 
        scale=alt.Scale(
            clamp=True, 
            domain=(60,90)
        )
    ),
    y=alt.Y(
        'value', 
        title="Value", 
        scale=alt.Scale(
            clamp=True, 
            domain=(3,7.5)
        )
    ),
    color=alt.Color(
        'variable', 
        title=None,
        legend=alt.Legend(orient="right"),
        scale=alt.Scale(range=custom_point_colors)
    ),
)
# points
points = base.mark_point(
    size=100,
    opacity=1,
)
# smooth
smooth = base.transform_loess(
    on='index', 
    loess='value', 
    groupby=['variable'], 
    bandwidth=0.06,
).mark_line(
    size=5,
    strokeDash=[10,7],
)
# chart
chart = alt.layer(
    points, smooth
).properties(
    title={
        "text" : "Altair Line Smoothing Plot Title",  
        "subtitle" : "Subtitle Text",
        "offset" : 15
    },
    width=width, 
    height=height
).interactive()
# display
chart.display()
```

```python
# regression plot
base = alt.Chart(
    data=df_iris
).mark_circle(
).encode(
    x=alt.X(
        "sepal_length", 
        scale=alt.Scale(zero=False)
    ), 
    y=alt.Y(
        "petal_width", 
        scale=alt.Scale(zero=False)
    ),
)
# regressions
regression_order_list = [1,2,4]
custom_line_colors = [
    colors_tableau['GREEN'], 
    colors_tableau['RED'], 
    colors_tableau['ORANGE']
]
regressions = []
for each_order in regression_order_list:
    regressions.append(
        base.transform_regression(
            on="sepal_length", 
            regression="petal_width", 
            method="poly", 
            order=each_order, 
            as_=["sepal_length", str(each_order)],
        ).mark_line(
            size=5,
            strokeDash=[25,10],
        ).transform_fold(
            [str(each_order)], 
            as_=["order", "petal_width"],
        ).encode(
            color=alt.Color(
                "order:N", 
                legend=None, 
                scale=alt.Scale(range=custom_line_colors)
            ),
        )
    )
# chart
chart = alt.layer(
    base, *regressions
).properties(
    title={
        "text" : "Altair Regression Plot Title",  
        "subtitle" : "Subtitle Text",
        "offset" : 15
    },
    width=width, 
    height=height
).interactive()
# display
chart.display()
```
