```python
# figure size and aspect ratio
golden_ratio = (1 + np.sqrt(5)) / 2
width = 12
height = width / golden_ratio
# fonts
font_size = 16
font_larger_size = 18
subtitle_font_size = 20
title_font_size = 36
# colors
color_font_axes_labels = colors["GRAY_4"]
color_title = colors["GRAY_3"]
color_subtitle = colors["GRAY_3"]
color_prop_cycle = matplotlib.cycler(color=color_cycle_custom)
# custom matplotlib parameters
custom_rc= {
    # figure 
    "figure.figsize" : [width, height],
    # fonts 
    "font.size" : font_size,
    # Arial 
    "font.family" : "Arial",
    "mathtext.fontset" : "custom",
    "mathtext.bf" : "Arial:bold",
    "mathtext.it" : "Arial:italic",
    # Helvetica (note: there is a bug with bold weight)
    #"font.family" : "Helvetica Neue",
    #"mathtext.fontset" : "custom",
    #"mathtext.bf" : "Helvetica Neue:bold",
    #"mathtext.it" : "Helvetica Neue:italic",
    # background colors
    "axes.facecolor" : "white",
    "patch.edgecolor": "white",
    # axes and labels colors 
    "text.color" : color_font_axes_labels,
    "xtick.color" : color_font_axes_labels,
    "ytick.color" : color_font_axes_labels,
    "axes.labelcolor" : color_font_axes_labels,
    "axes.edgecolor" : color_font_axes_labels,
    # plot color cycle
    "axes.prop_cycle" : color_prop_cycle,
    # grid
    "axes.grid": False,
    # ticks
    "xtick.bottom": True,
    "ytick.left": True,
    "xtick.direction": "out",
    "ytick.direction": "out",
    "xtick.major.size" : 5,
    "xtick.minor.size" : 2.5,
    "ytick.major.size" : 5,
    "ytick.minor.size" : 2.5,
    "xtick.major.width" : .8,
    "xtick.minor.width" : .8,
    "ytick.major.width" : .8,
    "ytick.minor.width" : .8,
    "xtick.major.pad" : 8,
    "xtick.minor.pad" : 8,
    "ytick.major.pad" : 8,
    "ytick.minor.pad" : 8,
    # tick labels 
    "xtick.labelsize" : font_size,
    "ytick.labelsize" : font_size,
    # axes
    "axes.axisbelow": False,
    "axes.spines.top" : False,
    "axes.spines.right": False,
    "axes.linewidth" : .8,
    "axes.labelpad" : 8,
    # axes labels 
    "axes.labelsize" : font_larger_size,
    # patches
    "patch.force_edgecolor": True,
    "patch.linewidth" : 1,
    # legend
    "legend.frameon" : False,
    "legend.fontsize" : font_larger_size,
    # lines and points
    "lines.solid_capstyle": "round",
    "lines.linewidth" : 3,
    "lines.markersize" : 10,
}
for each in custom_rc:
    matplotlib.rcParams[each] = custom_rc[each]
```

```python
# histogram
fig, ax = plt.subplots()
ax.hist(
    x=df_iris["sepal_width"], 
    alpha=0.75, 
    bins=20)
# text 
ax.text(
    x=0.058, y=.988, 
    s='Matplotlib Histogram Title', 
    fontweight="bold",
    fontsize=title_font_size, 
    color=color_title, 
    transform=fig.transFigure
)
ax.text(
    x=0.06, y=.922, 
    s="Subtitle Text", 
    fontsize=subtitle_font_size, 
    color=color_subtitle, 
    transform=fig.transFigure
)
plt.xlabel("sepal_width")
plt.ylabel("Number")
# display
plt.show()
plt.close()
```

```python
# bar plot
fig, ax = plt.subplots()
bar_data = df_iris.mean()
bar_index = np.arange(len(bar_data))
bar_height = 0.7
bars = ax.barh(
    y=bar_index, width=bar_data, 
    height=bar_height, 
    alpha=0.9
)
ax.invert_yaxis()
# text
ax.text(
    x=-0.046, y=.955,
    s="Matplotlib Bar Title",
    fontweight="bold",
    fontsize=title_font_size, 
    color=color_title, 
    transform=fig.transFigure
)
ax.text(
    x=-0.044, y=.888,
    s="Subtitle Text", 
    fontsize=subtitle_font_size,
    color=color_subtitle, 
    transform=fig.transFigure
)
ax.spines['left'].set_visible(False)
ax.tick_params(axis='y', length=0)
# text annotations 
for each_index, each_bar in enumerate(bars):
    each_value = np.round(df_iris.mean()[each_index], decimals=3)
    plt.text(
        x=each_bar.get_width()-0.7, 
        y=each_bar.get_y()+.36, 
        s=each_value, 
        fontsize=24, 
        color="white", 
        weight="bold", 
        va="center"
    )
plt.yticks(bar_index, df_iris.columns + " mean")
plt.xlabel("Value")
# display
plt.show()
plt.close()
```

```python
# grouped bar
fig, ax = plt.subplots()
bar_data = df_iris.groupby("species").mean()
bar_index = np.arange(bar_data.shape[1])
bar_width = 0.7
for each_index in range(bar_data.shape[0]):
    ax.bar(
        x=bar_index - bar_width/2 + (each_index / bar_data.shape[0] * bar_width),
        height=bar_data.iloc[each_index], 
        width=bar_width/bar_data.shape[0], 
        alpha=0.75, 
        align='edge'
    )
ax.legend(df_iris.species.unique(), loc=(1,.8))
# text
ax.text(
    x=0.065, y=.99, 
    s="Matplotlib Grouped Bar Title", 
    fontweight="bold",
    fontsize=title_font_size, 
    color=color_title,
    transform=fig.transFigure
)
ax.text(
    x=0.068, y=.925,
    s="Subtitle Text", 
    fontsize=subtitle_font_size, 
    color=color_subtitle, 
    transform=fig.transFigure
)
plt.xticks(bar_index, df_iris.columns[0:4] + " mean")
plt.xlabel("Feature")
plt.ylabel("Value")
# display
plt.show()
plt.close()
```

```python
# grouped bar
bar_data = df_iris.groupby("species").mean().T
custom_bar_colors = [
    colors_tableau['BLUE'], 
    colors['GRAY_7'], 
    colors['GRAY_9']
]
ax = bar_data.plot(
    kind='bar', 
    color=custom_bar_colors, 
    width=.7, 
    alpha=0.9
)
# text 
plt.figtext(
    x=0.065, y=.99, 
    s="Matplotlib Grouped Bar Title", 
    fontweight="bold",
    fontsize=title_font_size,
    color=color_title
)
plt.figtext(
    x=0.068, y=.925, 
    s="Subtitle Text", 
    fontsize=subtitle_font_size, 
    color=color_subtitle
)
plt.legend(loc=(1,.8))
plt.xlabel("Feature")
plt.ylabel("Value")
plt.xticks(rotation=0) 
# text annotations 
for each_index, each_bar in enumerate(ax.patches):
    if each_index in range(0,4):
        each_value = round(each_bar.get_height(), ndigits=1)
        ax.text(
            x=each_bar.get_x() + .03, 
            y=each_bar.get_height() + 0.15,
            s=each_value, 
            fontsize=20, 
            color=colors['GRAY_1'],
            weight="bold", 
            va="center"
        )
# display
plt.show()
plt.close()
```

```python
# datashader
number_of_points = 10_000_000
random_points = np.random.multivariate_normal((0,0), [[0.5,0.5], [0.5,10.0]], (number_of_points,))
df_random_points = pd.DataFrame(random_points)
df_random_points.columns = ['x','y']
# canvas
cvs = ds.Canvas(
    plot_width=400, 
    plot_height=400,
)
# aggregation 
agg = cvs.points(
    df_random_points, 
    'x', 'y'
)
# image
img = tf.set_background(
    tf.shade(agg), 
    color='white'
)
```

```python
# heatmap
num_xticks = 7
num_yticks = 7
xticks = np.linspace(0, len(img.x.values) - 1, num_xticks, dtype=np.int)
yticks = np.linspace(0, len(img.y.values) - 1, num_yticks, dtype=np.int)
xticklabels = [int(img.x.values[idx]) for idx in xticks]
yticklabels = [int(img.y.values[idx]) for idx in yticks]
plt.figure(figsize = (8,8))
ax = sns.heatmap(
    img.values,
    xticklabels=xticklabels,
    yticklabels=yticklabels,
    cmap="rainbow_r",
    square=True,
    cbar=False,
)
ax.set_xticks(xticks)
ax.set_yticks(yticks)
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
ax.invert_yaxis()
ax.set_ylabel('y') 
ax.set_xlabel('x') 
# text 
plt.figtext(
    x=0.03, y=.98, 
    s="Seaborn Datashader Heatmap", 
    fontweight="bold",
    fontsize=30,
    color=color_title
)
plt.figtext(
    x=0.032, y=.925,
    s="Subtitle Text", 
    fontsize=20, 
    color=color_subtitle, 
)
plt.show()
plt.close()
```
