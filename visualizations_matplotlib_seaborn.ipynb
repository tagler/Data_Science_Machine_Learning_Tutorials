```python
# figure size and aspect ratio
golden_ratio = (1 + np.sqrt(5)) / 2
width = 12
height = width / golden_ratio
# fonts
font_size = 16
font_larger_size = 18
subtitle_font_size = 20
title_font_size = 36
# colors
custom_font_axes_labels_color = custom_colors["GRAY_4"]
custom_title_color = custom_colors["GRAY_3"]
custom_subtitle_color = custom_colors["GRAY_3"]
color_cycle = [
    custom_colors["BLUE_1"],
    custom_colors["GREEN_1"],
    custom_colors["RED_1"],
    custom_colors["ORANGE_1"],
    custom_colors["PURPLE_1"],
]
color_cycle_2 = [
    custom_colors["GRAY_5"],
    custom_colors["GRAY_6"],
    custom_colors["GRAY_7"],
    custom_colors["GRAY_8"],
    custom_colors["GRAY_9"],
]
custom_color_cycle = matplotlib.cycler(color=color_cycle)
# custom matplotlib parameters
custom_rc= {
    # figure 
    "figure.figsize" : [width, height],
    # fonts 
    "font.size" : font_size,
    # Arial 
    "font.family" : "Arial",
    "mathtext.fontset" : "custom",
    "mathtext.bf" : "Arial:bold",
    "mathtext.it" : "Arial:italic",
    # Helvetica (note: there is a bug with bold weight)
    #"font.family" : "Helvetica Neue",
    #"mathtext.fontset" : "custom",
    #"mathtext.bf" : "Helvetica Neue:bold",
    #"mathtext.it" : "Helvetica Neue:italic",
    # background colors
    "axes.facecolor" : "white",
    "patch.edgecolor": "white",
    # axes and labels colors 
    "text.color" : custom_font_axes_labels_color,
    "xtick.color" : custom_font_axes_labels_color,
    "ytick.color" : custom_font_axes_labels_color,
    "axes.labelcolor" : custom_font_axes_labels_color,
    "axes.edgecolor" : custom_font_axes_labels_color,
    # plot color cycle
    "axes.prop_cycle" : custom_color_cycle,
    # grid
    "axes.grid": False,
    # ticks
    "xtick.bottom": True,
    "ytick.left": True,
    "xtick.direction": "out",
    "ytick.direction": "out",
    "xtick.major.size" : 5,
    "xtick.minor.size" : 2.5,
    "ytick.major.size" : 5,
    "ytick.minor.size" : 2.5,
    "xtick.major.width" : .8,
    "xtick.minor.width" : .8,
    "ytick.major.width" : .8,
    "ytick.minor.width" : .8,
    "xtick.major.pad" : 8,
    "xtick.minor.pad" : 8,
    "ytick.major.pad" : 8,
    "ytick.minor.pad" : 8,
    # tick labels 
    "xtick.labelsize" : font_size,
    "ytick.labelsize" : font_size,
    # axes
    "axes.axisbelow": False,
    "axes.spines.top" : False,
    "axes.spines.right": False,
    "axes.linewidth" : .8,
    "axes.labelpad" : 8,
    # axes labels 
    "axes.labelsize" : font_larger_size,
    # patches
    "patch.force_edgecolor": True,
    "patch.linewidth" : 1,
    # legend
    "legend.frameon" : False,
    "legend.fontsize" : font_larger_size,
    # lines and points
    "lines.solid_capstyle": "round",
    "lines.linewidth" : 3,
    "lines.markersize" : 10,
}
for each in custom_rc:
    matplotlib.rcParams[each] = custom_rc[each]
```

```python
# barh
fig, ax = plt.subplots()
bar_data = df_iris.mean()
bar_index = np.arange(len(bar_data))
bar_height = 0.7
bars = ax.barh(y=bar_index, width=bar_data, height=bar_height)
ax.invert_yaxis()
ax.text(-0.046, .955, "Matplotlib Bar Title", fontweight="bold",
         fontsize=title_font_size, color=custom_title_color, transform=fig.transFigure)
ax.text(-0.044, .888, "Subtitle Text", 
         fontsize=subtitle_font_size, color=custom_subtitle_color, transform=fig.transFigure)
ax.spines['left'].set_visible(False)
ax.tick_params(axis='y', length=0)
for each_index, each_bar in enumerate(bars):
    each_value = np.round(df_iris.mean()[each_index], decimals=3)
    plt.text(each_bar.get_width()-0.7, each_bar.get_y()+.36, 
             each_value, fontsize=24, color="white", weight="bold", va="center")
plt.yticks(bar_index, df_iris.columns + " mean")
plt.xlabel("Value")
plt.show()
plt.close()
```

```python
# data
raw_data = [[40,9],[25,30],[19,38],[11,13],[4,5]]
processed_data = pd.DataFrame()
for each_index, each_data in enumerate(raw_data):
    each_df = pd.DataFrame(each_data).reset_index()
    each_df['LABEL'] = chr(ord('@')+each_index+1)
    each_df.columns = ["TIME","VALUE","LABEL"]
    processed_data = pd.concat([processed_data, each_df]).reset_index(drop=True)
# slopeplot
fig, ax = plt.subplots(figsize=(8,6))
for each_group in list(processed_data.LABEL.unique()):
    # colors
    if each_group == 'A':
        color = custom_colors['RED_1']
    elif each_group in ['B','C']:
        color = custom_colors['BLUE_1']
    else:
        color = custom_colors['GRAY_5']
    each_df = processed_data[processed_data['LABEL'] == each_group]
    ax.plot(each_df.TIME, each_df.VALUE, 'o-', alpha=0.75, color=color,
            linewidth=5, markersize=10)
# text labels
for each_point in processed_data.itertuples():
    # colors
    if each_point.LABEL == 'A':
        color = custom_colors['RED_1']
    elif each_point.LABEL in ['B','C']:
        color = custom_colors['BLUE_1']
    else:
        color = custom_colors['GRAY_5']
    # left 
    if each_point.TIME == 0:
        plt.annotate(f"{each_point.VALUE}%", 
                     [each_point.TIME - .05, each_point.VALUE - .8],
                     fontsize=font_larger_size, fontweight='bold', ha='right', color=color)
    # right 
    elif each_point.TIME == 1:
        plt.annotate(f"{each_point.VALUE}%", 
                     [each_point.TIME + .055 , each_point.VALUE - .8],
                     fontsize=font_larger_size, fontweight='bold', ha='left', color=color)
        plt.annotate(f"Option {each_point.LABEL}", 
                     [each_point.TIME + .2 , each_point.VALUE - .8],
                     fontsize=font_larger_size, fontweight='bold', ha='left', color=color)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(True)
ax.spines['bottom'].set_bounds(0, 1)
ax.tick_params(axis='y', length=0)
ax.tick_params(axis='x', length=5)
# titles
ax.text(0.14, .997, "Matplotlib Slopeplot Title", fontweight="bold",
         fontsize=title_font_size, color=custom_title_color, transform=fig.transFigure)
ax.text(0.142, .92, "Subtitle Text", 
         fontsize=subtitle_font_size, color=custom_subtitle_color, transform=fig.transFigure)
# text annotations
ax.text(1.08, 0.78, "BEFORE:\ntext event description          ", 
         fontsize=font_larger_size, color=custom_subtitle_color, transform=fig.transFigure)
ax.text(1.08, 0.62, "AFTER:\ntext event description          ", 
         fontsize=font_larger_size, color=custom_subtitle_color, transform=fig.transFigure)
plt.xticks(ticks=[0, 1], labels=['BEFORE', 'AFTER'])
plt.yticks(ticks=[], labels=[])
plt.xlim(-0.2,1.2)
plt.ylim(0.5,41)
plt.xlabel("")
plt.ylabel("")
plt.show()
plt.close()
```
