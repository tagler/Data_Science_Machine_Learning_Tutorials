```python
# note: use matplotlib for titles due to bugs in plotnine with plotting outside of plot area\
# reference: https://github.com/has2k1/plotnine/issues/290
```

```python
# matplotlib settings
custom_rc= {
    # Arial 
    "font.family" : "Arial",
    "mathtext.fontset" : "custom",
    "mathtext.bf" : "Arial:bold",
    "mathtext.it" : "Arial:italic",
    # Helvetica (note: bug with bold weight)
    #"font.family" : "Helvetica Neue",
    #"mathtext.fontset" : "custom",
    #"mathtext.bf" : "Helvetica Neue:bold",
    #"mathtext.it" : "Helvetica Neue:italic",
}
for each in custom_rc:
    matplotlib.rcParams[each] = custom_rc[each]
```

```python
# size
golden_ratio = (1 + np.sqrt(5)) / 2
width = 8
height = width / golden_ratio
plotnine.options.figure_size = (width, height)
# fonts
font_size = 12
font_larger_size = 13
subtitle_font_size = 14
title_font_size = 22
font_family = "Helvetica Neue"
# colors
color_font_axes_labels = colors["GRAY_4"]
color_title = colors["GRAY_3"]
color_subtitle = colors["GRAY_3"]
```

```python
custom_theme = (
    theme_classic(
        base_size=font_size,
        base_family=font_family
    )
    + theme(
        panel_border=element_blank(),
        # background
        plot_background=element_rect(
            color=colors['WHITE_1']
        ),
        plot_margin=None,
        # axis and ticks
        axis_line=element_line(
            color=color_font_axes_labels, 
            size=.5
        ),
        axis_ticks=element_line(
            color=color_font_axes_labels, 
            size=.5
        ),
        axis_text=element_text(
            color=color_font_axes_labels, 
            size=font_size,
            family=font_family, 
            margin={'t': 5, 'r': 5}
        ),
        axis_title=element_text(
            color=color_font_axes_labels,
            size=font_larger_size,
            family=font_family, 
            margin={'t': 6,'r': 7}
        ),
        axis_ticks_length_major=3,
        axis_ticks_length_minor=1.5,
        # title
        title=element_text(
            color=color_title, 
            size=title_font_size, 
            family=font_family
        ),
        # text
        text=element_text(
            color=color_font_axes_labels, 
            size=font_size, 
            family=font_family
        ),
        # grid
        panel_grid_major=element_blank(),
        panel_grid_minor=element_blank(),
        panel_grid_major_x=element_blank(),
        panel_grid_major_y=element_blank(),
        panel_grid_minor_x=element_blank(),
        panel_grid_minor_y=element_blank(),
        # legend
        legend_key=element_blank(),
        legend_text=element_text(
            color=color_font_axes_labels, 
            size=font_size,
            family=font_family, 
            fontweight="normal", 
            margin={'l': 8}
        ),
        legend_title=element_text(
            color=color_font_axes_labels, 
            size=font_size, 
            family=font_family, 
            fontweight="bold", 
            margin={'b': 10}
        ),
    )
)
```

```python
# histogram
figure = (
    ggplot(
        data=df_iris, 
        mapping=aes(x='sepal_width')
    )
    + geom_histogram(
        bins=20, 
        size=0.25, 
        fill=colors_tableau["BLUE"], 
        color=colors['WHITE_1'], 
        alpha=0.75
    )
    + scale_x_continuous(
        minor_breaks=[]
    ) 
    + scale_y_continuous(
        minor_breaks=[], 
        expand=(0,0)
    ) 
    + labs(
        x = "sepal_width", 
        y = "Number"
    )
    + custom_theme
)
# add text
ax = figure.draw().get_axes()
plt.figtext(
    x=0.057, y=.992, 
    s="ggplot Histogram Title", 
    fontweight="bold", 
    fontsize=title_font_size, 
    color=color_title
)
plt.figtext(
    x=0.059, y=.922, 
    s="Subtitle Text", 
    fontsize=subtitle_font_size, 
    color=color_subtitle
)
# display
plt.show()
plt.close()
```

```python
# histogram
figure = (
    ggplot(
        data=df_iris, 
        mapping=aes(x='sepal_width')
    )
    + geom_histogram(
        aes(y='stat(density)'), 
        bins=20, 
        size=0.25, 
        fill=colors_tableau["BLUE"], 
        color=colors['WHITE_1'], 
        alpha=0.75
    )
    + geom_density(
        color=colors["BLUE_1"], 
        fill=colors_tableau["BLUE"],
        alpha=0.5,
        size=1
    )
    + scale_x_continuous(
        minor_breaks=[]
    ) 
    + scale_y_continuous(
        breaks=np.arange(0,2,0.2), 
        minor_breaks=[], 
        expand=(0,0)
    ) 
    + labs(
        x = "sepal_width", 
        y = "Number"
    ) 
    + custom_theme
)
# add text
ax = figure.draw().get_axes()
plt.figtext(
    x=0.05, y=.992, 
    s="ggplot Histogram Title", 
    fontweight="bold", 
    fontsize=title_font_size, 
    color=color_title
)
plt.figtext(
    x=0.052, y=.922, 
    s="Subtitle Text", 
    fontsize=subtitle_font_size,
    color=color_subtitle
)
# display
plt.show()
plt.close()
```

```python
# histogram
df_iris_melt = df_iris.melt(
    value_vars=["sepal_length", "sepal_width", "petal_length", "petal_width"]
)
df_iris_melt.columns = ['Feature','Value']
figure = (
    ggplot(
        data=df_iris_melt, 
        mapping=aes(x='Value', fill='Feature')
    )
    + geom_histogram(
        bins=20, 
        size=0.25, 
        color=colors['WHITE_1'], 
        alpha=0.8
    ) 
    + scale_fill_manual(color_cycle_custom)
    + scale_x_continuous(
        breaks=np.arange(0,10,2), 
        minor_breaks=[]
    )
    + scale_y_continuous(
        minor_breaks=[], 
        expand=(0,0)
    ) 
    + labs(
        x = "Value", 
        y = "Number", 
        fill = ""
    )
    + custom_theme
    + theme(legend_position = (1,.8))
)
# add text
ax = figure.draw().get_axes()
plt.figtext(
    x=0.057, y=1.01, 
    s="ggplot Multiple Histogram Title", 
    fontweight="bold",
    fontsize=title_font_size, 
    color=color_title
)
plt.figtext(
    x=0.059, y=.938, 
    s="Subtitle Text",
    fontsize=subtitle_font_size, 
    color=color_subtitle
)
plt.show()
plt.close()
```

```python
# polyregression smoothing 
def poly(x, degree=1):
    d = {}
    for i in range(degree+1):
        if i == 1:
            d['x'] = x
        else:
            d[f'x**{i}'] = np.power(x, i)
    return pd.DataFrame(d)
figure = (
    ggplot(
        data=df_iris, 
        mapping=aes(x='sepal_length', y='petal_width')
    )
    + geom_point(
        alpha=1, 
        size=3, 
        color=colors_tableau["BLUE"]
    )
    + geom_smooth(
        method='lm', 
        fill=colors_tableau["GREEN"], 
        color=colors_tableau["GREEN"], 
        level=.99
    )
    + stat_smooth(
        method='lm', 
        formula='y ~ poly(x, degree=2)', 
        color=colors["RED_1"], 
        fill=colors["RED_1"], 
        se=True, 
        level=.99
    )
    + scale_x_continuous(minor_breaks=[])
    + scale_y_continuous(minor_breaks=[])
    + labs(
        x = "sepal_length", 
        y = "petal_width"
    )
    + custom_theme
)
# add text
ax = figure.draw().get_axes()
plt.figtext(
    x=0.065, y=.995, 
    s="ggplot Regression Plot Title", 
    fontweight="bold",
    fontsize=title_font_size, 
    color=color_title
    
)
plt.figtext(
    x=0.067, y=.925, 
    s="Subtitle Text",
    fontsize=subtitle_font_size, 
    color=color_subtitle
)
# display
plt.show()
plt.close()
```

```python
# heatmap
df_heatmap = df_iris.corr()
df_heatmap = df_heatmap.replace(1, np.NaN)
df_heatmap = df_heatmap.where(np.triu(np.ones(df_heatmap.shape)).astype(np.bool))
df_heatmap_melt = df_heatmap.reset_index().melt(
    id_vars=['index'], 
    value_vars=["sepal_length", "sepal_width", "petal_length","petal_width"]
)
df_heatmap_melt.columns = ['variable_x', 'variable_y', 'value']
df_heatmap_melt = df_heatmap_melt[df_heatmap_melt['value'].notnull()]
# workaround for feature order in plot
category_order = list(df_heatmap.index)[::-1]
df_heatmap_melt_ordered = df_heatmap_melt.copy()
df_heatmap_melt_ordered['variable_x'] = pd.Categorical(
    df_heatmap_melt_ordered['variable_x'], 
    categories=category_order[::-1], 
    ordered=True
)
df_heatmap_melt_ordered['variable_y'] = pd.Categorical(
    df_heatmap_melt_ordered['variable_y'], 
    categories=category_order, 
    ordered=True
)
figure = (
    ggplot(
        data=df_heatmap_melt_ordered, 
        mapping=aes(x='variable_x', y='variable_y', fill='value')
    )
    + geom_tile(
        aes(width=.98, height=.98), 
        color=colors['WHITE_1']
    )
    + scale_fill_continuous(
        "coolwarm", 
        limits=[-1,1]
    )
    + labs(
        x = "", 
        y = "", 
        fill=""
    )
    + custom_theme
    + theme(
        axis_line=element_blank(),
        axis_ticks=element_blank(),
        axis_text=element_text(margin={'t': 2, 'r': 2}),
        figure_size=(5,5)
    )
)
# add text
ax = figure.draw().get_axes()
plt.figtext(
    x=-.08, y=.965, 
    s="ggplot Correlation Plot Title", 
    fontweight="bold",
    fontsize=title_font_size, 
    color=color_title
)
plt.figtext(
    x=-.082, y=.89, 
    s="Subtitle Text",
    fontsize=subtitle_font_size, 
    color=color_subtitle
)
# display
plt.show()
plt.close()
```
